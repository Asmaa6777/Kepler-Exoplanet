```python
# non-useful and data leakage columns
non_inputs = [
    'kepid', 'kepoi_name', 'kepler_name',
    'koi_score', 'koi_tce_delivname', 'koi_fpflag_nt',
    'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec', 'dec', 'ra'
]
non_inputs.extend([col for col in df.columns if '_err' in col])

df = df.drop(non_inputs, axis=1)
```

```python
df.info()
```

```python
sns.histplot(data=df, x='koi_kepmag', bins=5)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='koi_disposition', y='koi_kepmag')
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='koi_disposition', y='koi_period')
```

```python
sns.pairplot(df, hue='koi_disposition')
```

```python

correlation_matrix = df.corr(numeric_only=True)
```

```python

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
